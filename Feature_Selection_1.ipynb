```python
import pandas as pd
data1=pd.read_csv("diabetes.csv")
data1.head()
```

```python
data1.isna().sum()
```

```python
miss_percent1=(data1.isnull().sum()/len(data1))*100
miss_percent1
```

```python
df_percent2=pd.DataFrame({'column name':data1.columns,'missing percent':miss_percent1.values})
df_percent2
```

```python
drop_percent1=df_percent2[df_percent2['missing percent']>30]['column name'].values
drop_percent1
```

```python
data1_dropped=data1.drop(drop_percent1,axis=1)
data1_dropped.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X1 = data1.drop(columns=['Outcome'])
X1_changed = data1_dropped.drop(columns=['Outcome'])
y1 = data1['Outcome']

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=42)
X1_changed_train, X1_changed_test, y1_changed_train, y1_changed_test = train_test_split(X1_changed, y1, test_size=0.3, random_state=42)

model1 = RandomForestClassifier(random_state=42)
model1.fit(X1_train, y1_train)
y1_pred = model1.predict(X1_test)
original_accuracy1 = accuracy_score(y1_test, y1_pred)
print("\nAccuracy with original feature set:", original_accuracy1)

model_reduced1 = RandomForestClassifier(random_state=42)
model_reduced1.fit(X1_changed_train, y1_changed_train)
y1_changed_pred = model_reduced1.predict(X1_changed_test)
changed_accuracy1 = accuracy_score(y1_changed_test, y1_changed_pred)
print("Accuracy with reduced feature set:", changed_accuracy1)

accuracy_change1 = original_accuracy1 - changed_accuracy1
print("\nAccuracy difference (Original - Reduced):", accuracy_change1)
```

```python
data2=pd.read_csv("melbourne.csv")
data2.head()
```

```python
data2=data2.drop(['Suburb','Type','Method','SellerG','Date','Postcode','Lattitude','Longitude','CouncilArea','Regionname'],axis=1,errors='ignore')
data2.head()
```

```python
data2.isna().sum()
```

```python
miss_percent2=(data2.isnull().sum()/len(data2))*100
miss_percent2
```

```python
df_percent2=pd.DataFrame({'column name':data2.columns,'miss percent':miss_percent2.values})
df_percent2
```

```python
drop_percent2=df_percent2[df_percent2['miss percent']>22]['column name'].values
drop_percent2
```

```python
data2_dropped=data2.drop(drop_percent2,axis=1)
data2_dropped.head()
```

```python
data2_dropped.columns
```

```python
# Imputer for features (X2)
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X2 = data2.drop(columns=['Price'])
X2_imputed = imputer.fit_transform(X2)

# Imputer for reduced features (X2_changed)
X2_changed = data2_dropped.drop(columns=['Price'])
X2_changed_imputed = imputer.fit_transform(X2_changed)

# Target variable (y2)
y2 = data2['Price']
y2_imputed = imputer.fit_transform(y2.values.reshape(-1, 1)).ravel()  # Ensure target variable is 1D

# Train-test split
X2_train, X2_test, y2_train, y2_test = train_test_split(X2_imputed, y2_imputed, test_size=0.3, random_state=42)
X2_changed_train, X2_changed_test, y2_changed_train, y2_changed_test = train_test_split(X2_changed_imputed, y2_imputed, test_size=0.3, random_state=42)


from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Train the model with original features using RandomForestRegressor
model2 = RandomForestRegressor(random_state=42)
model2.fit(X2_train, y2_train)
y2_pred = model2.predict(X2_test)

# Evaluate using RMSE and R score for original feature set
original_rmse2 = mean_squared_error(y2_test, y2_pred, squared=False)
original_r2 = r2_score(y2_test, y2_pred)
print("\nRMSE with original feature set:", original_rmse2)
print("R with original feature set:", original_r2)

# Train the model with reduced features using RandomForestRegressor
model_reduced2 = RandomForestRegressor(random_state=42)
model_reduced2.fit(X2_changed_train, y2_changed_train)
y2_changed_pred = model_reduced2.predict(X2_changed_test)

# Evaluate using RMSE and R score for the reduced feature set
changed_rmse2 = mean_squared_error(y2_changed_test, y2_changed_pred, squared=False)
changed_r2 = r2_score(y2_changed_test, y2_changed_pred)
print("RMSE with reduced feature set:", changed_rmse2)
print("R with reduced feature set:", changed_r2)

# RMSE and R difference
rmse_change2 = original_rmse2 - changed_rmse2
r2_change2 = original_r2 - changed_r2
print("\nRMSE difference (Original - Reduced):", rmse_change2)
print("R difference (Original - Reduced):", r2_change2)

```
